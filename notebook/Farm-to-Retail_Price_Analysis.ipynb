```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
save_dir = './Desktop/Insight Project/Data/farm-to-retail-price/plots/'
```

```python
commodities = ['broccoli', 'iceberg_lettuce', 'potatoes', 'tomatoes']
```

```python
def read_csv_sort(productname):
    input_data = pd.read_csv('./Desktop/Insight Project/Data/farm-to-retail-price/'+productname+'.csv')
    
    # Date to Index
    input_data.index = input_data["Year"]
    del input_data["Year"]
    
    output_data = input_data
    
    return output_data
```

```python
# Test def read_csv_sort
read_csv_sort('broccoli')
```

```python
def plot_save_produce(productdata,productname):
    plt.rcParams['xtick.labelsize']=14
    plt.rcParams['ytick.labelsize']=14
    plt.figure(figsize=(10,5))
    plt.xticks(rotation=45)

    datelist = pd.to_datetime(productdata.index, format='%Y')
    plt.plot_date(datelist, productdata['Retail'], '-o')
    plt.plot_date(datelist, productdata['Farm'],'-o')
    
    graphtitle = productname
    
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Price (cents/lb)', fontsize=14)
    plt.title(graphtitle)
    plt.legend(loc='best')
    
    filetitle = productname
    plt.savefig(save_dir+filetitle+'.png', format='png', bbox_inches='tight')
    
    return plt.figure()
```

```python
pd.to_datetime(read_csv_sort('broccoli').index, format='%Y')
```

```python
test_data = read_csv_sort('broccoli')
fig_plot_save_produce = plot_save_produce(test_data,'broccoli')
```

```python
# loop over cities, commodities
for c in commodities:
    test_data = read_csv_sort(c)
    plot_save_produce(test_data, c)
```
